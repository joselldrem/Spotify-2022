```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
# get data
dataframe = pd.read_csv('C:\\Users\\Josell Remaneses\\Desktop\\Data Analysis\\own projects\\Spotify 2022\\spotify_top_charts_22_f.csv')
```

# Get Top 10 Tracks on Weekly Charts

```python
dataframe = dataframe.sort_values(by='weeks_on_chart',ascending=False)
dataframe_top10_tracks = dataframe[['track_name','weeks_on_chart']].head(10)
dataframe_top10_tracks 
```

```python
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']

plt.barh(dataframe_top10_tracks['track_name'] , dataframe_top10_tracks['weeks_on_chart'], color=colors)

plt.title('Top 10 Tracks on Charts 2022')
plt.xlabel('Weeks on Chart')
plt.ylabel('Track')

plt.show()


```

```python
dataframe_top10_artists = dataframe[['artist_names','weeks_on_chart']].head(10)
dataframe_top10_artists
```

```python
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']

plt.barh(dataframe_top10_artists['artist_names'] , dataframe_top10_artists['weeks_on_chart'], color=colors)

plt.title('Top 10 Artists on Charts 2022')
plt.xlabel('Weeks on Chart')
plt.ylabel('Artist')

plt.show()
```

```python
dataframe = dataframe.sort_values(by='duration_ms',ascending=False)
dataframe_top10_tracks = dataframe[['track_name','duration_ms']].head(10)
dataframe_top10_tracks.rename(columns={'duration_ms' : 'duration_mins'})
dataframe_top10_tracks['duration_ms'] = round(dataframe_top10_tracks['duration_ms']/60000,2)

dataframe_top10_tracks
```

```python
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']

plt.barh(dataframe_top10_tracks['track_name'] , dataframe_top10_tracks['duration_ms'], color=colors)

plt.title('Top 10 Longest TRacks 2022')
plt.xlabel('Duration(mins)')
plt.ylabel('Track')

plt.show()
```

```python
dataframe = dataframe.sort_values(by='danceability',ascending=False)
dataframe_top10_danceability = dataframe[['track_name','danceability']].head(10)
dataframe_top10_danceability
```

```python
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']

plt.barh(dataframe_top10_danceability['track_name'] , dataframe_top10_danceability['danceability'], color=colors)

plt.title('Top 10 Danceablet Tracks 2022')
plt.xlabel('Danceability Score')
plt.ylabel('Track')

plt.show()
```
